```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np

# Define RGB values for colors
colors = {
    'Orange (O)': [255, 165, 0],
    'DarkOrange (DO)': [255, 140, 0],
    'LightCoral (LC)': [250, 128, 128],
    'Red (R)': [255, 0, 0],
    'Aqua (A)': [0, 255, 255]
}

# colors = {
#     'Lawngreen': [124, 252, 0],
#     'Lime': [0, 255, 0],
#     'Pink': [255, 192, 203],
#     'Plum': [221, 160, 221],
#     'Red': [255, 0, 0]
# }

# Convert RGB values to numpy array
data = np.array(list(colors.values()))

# Perform hierarchical clustering using single linkage
linked = linkage(data, method='single')

# Create dendrogram plot
plt.figure(figsize=(10, 6))
dendrogram(linked,
           labels=list(colors.keys()),
           distance_sort='ascending',
           show_leaf_counts=True)

plt.title('Dendrogram for Selected Colors')
plt.xlabel('Color')
plt.ylabel('Euclidean Distance')
plt.tight_layout()
plt.show()
```
